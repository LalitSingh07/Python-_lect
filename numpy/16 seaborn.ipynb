## Plotting a Distplot

```python
import matplotlib.pyplot as plt

import seaborn as sns

sns.displot([10,20,30,40,10])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([0, 1, 2, 3, 4, 5], hist=False)

plt.show()
```
